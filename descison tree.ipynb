```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("train.csv").as_matrix()
```

```python
clf = DecisionTreeClassifier()
```

```python
xtrain = data[0:2000, 1:]
train_label = data[0:2000, 0]
```

```python
clf.fit(xtrain, train_label)
```

```python
xtest = data[2000:, 1:]
actual_label = data[2000:,0]
```

```python
d = xtest[29]
d.shape=(28,28)
pt.imshow(255-d, cmap='gray')
print(clf.predict([xtest[29]]))
pt.show()
```

```python
print(clf.score(xtest,actual_label))
```
